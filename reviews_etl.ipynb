```python
import pandas as pd
import os
from datetime import datetime as dt
```

```python
filepath = os.path.join('Data', 'All_airbnb_listings.csv')
df = pd.read_csv(filepath, header=0)
```

```python
df.head()
```

```python
df[df['reviews']<0]=0
```

```python
df["reviews"].min()
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
clean_data = df.loc[df['date'] > '2015-03-01']
```

```python
review_df= pd.DataFrame(clean_data.groupby(['date'])['reviews'].mean())
review_df.reset_index()
```

```python
review_df.to_csv("data/Reviews.csv",sep=",",index=True)
```

```python
price_data = df.loc[(df['price']<1000) &(df['price']>1)]
```

```python
price_data['price'].min()
```

```python
price_data
```

```python
price= pd.DataFrame(price_data.groupby(['date'])['price'].mean())
price.reset_index()
```

```python
price.to_csv("data/Price.csv", sep=',', index=True)
```

## Getting month of the date


```python
df['month'] = pd.DatetimeIndex(df['date']).month
df.head()
```

```python
review_month= pd.DataFrame(df.groupby(['month'])['price'].mean())
review_month.reset_index()
```

```python
df['day-of-week'] = df['date'].dt.day_name()
df.head()
```

```python
review_day= pd.DataFrame(df.groupby(['day-of-week'])['price'].mean())
review_day.reset_index()
```

```python
room_type= pd.DataFrame(df.groupby(['room_type'])['price'].count())
room_type.reset_index()
```
